```python
import tkinter as tk

root = tk.Tk()
root.geometry("850x750")
root.title("My CGPA Manager")

mark_s = []
c_hour = []
stu_record = []

# Code of menu_button for 1st screen display
def menu():
    # Creating frame 1 for displaying two buttons
    f1 = tk.Frame(root, bg="grey", width=850, height=750)
    f1.pack(fill="both", expand=True)
    # Adding labels in the frame 1 and packing them with place
    menu_label = tk.Label(f1, text="Welcome to our Result Manager", width=45, height=2, borderwidth=5, relief="ridge")
    menu_label_2 = tk.Label(f1, text="Select any button you want", width=45, height=2, borderwidth=5, relief="ridge")
    menu_label.place(x=250, y=90)
    menu_label_2.place(x=250, y=200)

    # Functionality for coordinator
    def cor_funct():
        f1.pack_forget()
        login_page()  # Calling login_page(), for coordinator login ID

    def login_page():
        f1.pack_forget()  # For removing function from the screen
        # Creating frame
        f3 = tk.Frame(root, bg="yellow", width=850, height=750)
        f3.pack(fill="both", expand=True)

        # Login and password for coordinator
        login_pass = "123"
        login_id = "123"

        # Creating labels in f3
        lab = tk.Label(f3, text="Enter your ID and password: ", width=45, height=2, borderwidth=5, relief="ridge")
        lab.place(x=250, y=90)

        user_lab = tk.Label(f3, text="Username: ", width=10, borderwidth=5, relief="ridge")
        user_lab.place(x=250, y=150)

        pass_lab = tk.Label(f3, text="Password: ", width=10, borderwidth=5, relief="ridge")
        pass_lab.place(x=250, y=210)

        # Use variable_class for storing user entered data in vari in f3
        username = tk.StringVar()
        password = tk.StringVar()

        # Now creating entry for password and userid and then packing it with place()
        name_en = tk.Entry(f3, textvariable=username, width=20, borderwidth=2, relief="ridge")
        name_en.place(x=356, y=152)

        pass_en = tk.Entry(f3, textvariable=password, width=20, borderwidth=2, relief="ridge")
        pass_en.place(x=356, y=212)

        def log_fuct():
            if password.get() == login_pass and username.get() == login_id:  # If password is correct then coordinator will enter record
                f3.pack_forget()
                record_stu()  # Calling record_stu(), so coordinator can add student record
            else:
                wron_lab = tk.Label(f3, text="Sorry Try Again: ", width=45, borderwidth=5, relief="ridge")
                wron_lab.place(x=250, y=450)

        def display_menu_again():  # To display the menu again
            f3.pack_forget()
            menu()

        # Adding button in the frame 3
        log_but = tk.Button(f3, text="Login", command=log_fuct, width=10, borderwidth=5, relief="sunken", bg="blue")
        log_but.place(x=550, y=300)

        again_menu_but = tk.Button(f3, text="Back", command=display_menu_again, width=7, borderwidth=5, relief="sunken", bg="blue")
        again_menu_but.place(x=70, y=60)

        def record_stu():
            f4 = tk.Frame(root, width=850, height=750, bg="green")
            f4.pack(fill="both", expand=True)

            # Creating labels in f4
            label_re = tk.Label(f4, text="Enter Student record here: ", width=25, borderwidth=5, relief="ridge")
            label_re.place(x=400, y=60)

            stu_name_lab = tk.Label(f4, text="Student Name:", width=15, borderwidth=5, relief="ridge")
            stu_name_lab.place(x=50, y=118)

            roll_no_lab = tk.Label(f4, text="Roll No:", width=15, borderwidth=5, relief="ridge")
            roll_no_lab.place(x=50, y=158)

            seme_lab = tk.Label(f4, text="Semester No.", width=15, borderwidth=5, relief="ridge")
            seme_lab.place(x=350, y=118)

            total_sub_lab = tk.Label(f4, text="Total subjects", width=15, borderwidth=5, relief="ridge")
            total_sub_lab.place(x=350, y=158)

            total_cgpa_lab = tk.Label(f4, text="Total CGPA", width=15, borderwidth=5, relief="ridge")
            total_cgpa_lab.place(x=50, y=195)

            # Storing entity data of student name, roll_no, semes using variable class
            stu_name = tk.StringVar()
            roll_no = tk.StringVar()
            seme_no = tk.StringVar()
            total_sub = tk.StringVar()
            total_cgpa = tk.StringVar()

            # Creating entry in f4
            en_stu_name = tk.Entry(f4, textvariable=stu_name, width=20, borderwidth=2, relief="ridge")
            en_stu_name.place(x=185, y=118)

            en_roll = tk.Entry(f4, textvariable=roll_no, width=20, borderwidth=2, relief="ridge")
            en_roll.place(x=185, y=158)

            en_seme = tk.Entry(f4, textvariable=seme_no, width=20, borderwidth=2, relief="ridge")
            en_seme.place(x=480, y=118)

            en_total_sub = tk.Entry(f4, textvariable=total_sub, width=20, borderwidth=2, relief="ridge")
            en_total_sub.place(x=480, y=158)

            en_total_cgpa = tk.Entry(f4, textvariable=total_cgpa, width=20, borderwidth=2, relief="ridge")
            en_total_cgpa.place(x=185, y=196)

            def next_function():
                fetch_sub_no = int(en_total_sub.get())
                fetch_seme_no = int(en_seme.get())
                current_cgpa = float(en_total_cgpa.get())

                f4.pack_forget()

                # Create f5
                f5 = tk.Frame(root, bg="grey", width=850, height=750)
                f5.pack(fill="both", expand=True)

                total_ch = tk.StringVar()

                marks_entries = []
                credit_hours_entries = []

                for i in range(1, fetch_sub_no + 1):
                    # Creating labels for marks and GPA and credit hours
                    marks_label = tk.Label(f5, text=f"Marks of subject {i}", width=20, borderwidth=5, relief="ridge")
                    marks_label.place(x=50, y=50 * i)

                    # Creating entry for marks
                    en_marks = tk.Entry(f5, width=14, borderwidth=2, relief="ridge")
                    en_marks.place(x=250, y=50 * i)
                    marks_entries.append(en_marks)

                    # Creating labels for credit hours
                    credit_hours_label = tk.Label(f5, text=f"Credit Hours of subject {i}", width=20, borderwidth=5, relief="ridge")
                    credit_hours_label.place(x=400, y=50 * i)

                    # Creating entry for credit hours
                    en_credit_hours = tk.Entry(f5, width=17, borderwidth=2, relief="ridge")
                    en_credit_hours.place(x=600, y=50 * i)
                    credit_hours_entries.append(en_credit_hours)

                # The total credit hour will display once and I want it to show after subject credit hours
                total_credit_hours_lab = tk.Label(f5, text="Total credit hours", width=18, borderwidth=2, relief="ridge")
                total_credit_hours = tk.Entry(f5, textvariable=total_ch, width=17, borderwidth=2, relief="ridge")
                # Packing label and entry for total credit hour
                total_credit_hours_lab.place(x=400, y=50 * (fetch_sub_no + 1))  # fetch_sub_no+1 will make the packing after credit hour entity and label
                total_credit_hours.place(x=600, y=50 * (fetch_sub_no + 1))

                def save_record():
                    # Fetch marks and credit hours data from the entry fields
                    marks = [int(entry.get()) for entry in marks_entries]
                    credit_hours = [int(entry.get()) for entry in credit_hours_entries]

                    # Calculate GPA
                    grade_points = []
                    for y in range(fetch_sub_no):
                        mark = marks[y]
                        if mark >= 81:
                            grade_point = 4.0
                        elif mark <= 49:
                            grade_point = 0.0
                        else:
                            grade_point = 4.0 - (80 - mark) * 0.1
                        grade_points.append(grade_point)

                    total = 0
                    for x in range(fetch_sub_no):
                        total += grade_points[x] * credit_hours[x]

                    new_cgpa = (total / sum(credit_hours))

                    # Calculate cumulative CGPA
                    total_credit_hours = sum(credit_hours)
                    cumulative_total_grade_points = new_cgpa * total_credit_hours

                    final_cgpa = cumulative_total_grade_points / total_credit_hours

                    # Store the student record
                    student_record = {
                        "name": stu_name.get(),
                        "roll_no": roll_no.get(),
                        "semester_no": fetch_seme_no,
                        "total_subjects": fetch_sub_no,
                        "total_cgpa": final_cgpa,
                        "marks": marks,
                        "credit_hours": credit_hours,
                        "total_credit_hours": int(total_ch.get())
                    }

                    # Append to student records list
                    stu_record.append(student_record)

                    # Write the data to a file
                    with open("student_records.txt", "a") as file:
                        file.write(str(student_record) + "\n")

                    f5.pack_forget()
                    final_option()  # After saving student record, moving to final screen

                save_button = tk.Button(f5, text="Save", command=save_record, width=10, borderwidth=5, relief="sunken", bg="blue")
                save_button.place(x=720, y=550)

            next_but = tk.Button(f4, text="Next", command=next_function, width=10, borderwidth=5, relief="sunken", bg="blue")
            next_but.place(x=600, y=258)

        def final_option():
            f6 = tk.Frame(root, bg="pink", width=850, height=750)
            f6.pack(fill="both", expand=True)
            # Adding labels in the f6
            lab_sel = tk.Label(f6, text="Select from the options ", width=45, borderwidth=5, relief="ridge")
            lab_sel.place(x=250, y=185)

            def display_menu_again():  # To display the menu again
                f6.pack_forget()
                menu()

            def add_more_student_record():  # To add more student records
                f6.pack_forget()
                log_fuct()

            def exiit():  # To end the program
                root.quit()

            # Creating buttons
            menu_again_but = tk.Button(f6, text="Display menu", command=display_menu_again, width=17, borderwidth=5, relief="sunken", bg="blue")
            menu_again_but.place(x=190, y=270)

            more_record = tk.Button(f6, text="Add more record", command=add_more_student_record, width=19, borderwidth=5, relief="sunken", bg="blue")
            more_record.place(x=440, y=270)

            exit_but = tk.Button(f6, text="Exit", command=exiit, width=9, borderwidth=5, relief="sunken", bg="blue")
            exit_but.place(x=680, y=270)

    # Functionality for student
    def stud_funct():
        f1.pack_forget()  # For removing function from the screen
        f2 = tk.Frame(root, bg="pink", width=850, height=750)
        f2.pack(fill="both", expand=True)

        # Adding labels in the frame
        lab1 = tk.Label(f2, text="Enter your roll no to see your GPA", width=45, height=2, borderwidth=5, relief="ridge")
        lab1.place(x=250, y=140)

        roll_lab = tk.Label(f2, text="Enter your roll no", width=15, height=2, borderwidth=5, relief="ridge")
        roll_lab.place(x=250, y=230)

        # Storing user-entered roll_no
        user_enter_roll = tk.StringVar()

        # Create entry
        show_roll_en = tk.Entry(f2, textvariable=user_enter_roll, width=10, borderwidth=5, relief="ridge")
        show_roll_en.place(x=380, y=235)

        def checking_iden():
            user_roll_no = user_enter_roll.get()
            found = False

            # Read student records from the file
            try:
                with open("student_records.txt", "r") as file:
                    for line in file:
                        record = eval(line.strip())
                        if record["roll_no"] == user_roll_no:
                            found = True
                            dis_lab = tk.Label(f2, text=f"Your CGPA: {record['total_cgpa']:.2f}", width=30, height=2, borderwidth=5, relief="ridge")
                            dis_lab.place(x=250, y=280)
                            break
            except FileNotFoundError:
                lab1 = tk.Label(f2, text="No data found", width=20, height=2, borderwidth=5, relief="ridge")
                lab1.place(x=250, y=280)

            if not found:
                lab = tk.Label(f2, text="Incorrect roll_no or No data found", width=30, height=2, borderwidth=5, relief="ridge")
                lab.place(x=250, y=280)

        def exiit():  # To end the program
            root.quit()

        def display_menu_again():  # To display the menu again
            f2.pack_forget()
            menu()

        # Create a button
        ok_but = tk.Button(f2, text="Ok", command=checking_iden, width=7, borderwidth=5, relief="sunken", bg="blue")
        ok_but.place(x=520, y=230)

        exit_but = tk.Button(f2, text="Exit", command=exiit, width=7, borderwidth=5, relief="sunken", bg="blue")
        exit_but.place(x=610, y=60)

        again_menu_but = tk.Button(f2, text="Back", command=display_menu_again, width=7, borderwidth=5, relief="sunken", bg="blue")
        again_menu_but.place(x=70, y=60)

    # Adding buttons in our frame of 1st screen
    stu_but = tk.Button(f1, text="Student", command=stud_funct, width=7, borderwidth=5, relief="sunken", bg="blue")
    cor_but = tk.Button(f1, text="Coordinator", command=cor_funct, width=9, borderwidth=5, relief="sunken", bg="blue")
    stu_but.place(x=300, y=300)
    cor_but.place(x=450, y=300)

menu()
root.mainloop()

 
```
